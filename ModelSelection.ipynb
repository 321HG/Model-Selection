```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression,LogisticRegression 
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from tkinter import *



```

```python
class ModelSelection:
    def __init__(self):
        self.dataset_name = 'data.csv'

        self.result = []
        
        self.dataset = pd.read_csv('data.csv')
        self.col = self.dataset.columns
        self.dataset_update()
        
    def dataset_update(self):
        self.x = self.dataset.iloc[:,1:-1]
        self.y = self.dataset.iloc[:,-1]
        self.x_train,self.x_test,self.y_train,self.y_test = train_test_split(self.x,self.y,test_size = 0.25)
        self.dataset.head()
         
    def SVM(self):
        self.model4 = SVC(kernel = 'linear')
        self.model4.fit(self.x_train,self.y_train)
        print(self.model4.score(self.x_test,self.y_test))
        l = ['Deepak',100,1,1,89,98]
        a = self.model4.predict([l[1:]])
        print(a,type(a))
        a = a[0]
        print(a,type(a))
        if a == 1.0:
            print("yes")
        elif a== 0:
            print("No")
        self.result.append(a)
        
        l.append(a)    
        print(l)
        self.updateDataset(l)        
    
    
    def RandomForest(self):
        self.model2 = RandomForestClassifier(n_estimators=5)
        self.model2.fit(self.x_train,self.y_train)
        print(self.model2.score(self.x_test,self.y_test))
        l = ['Deepak',100,1,1,65,98]
        a = self.model2.predict([l[1:]])
        print(a,type(a))
        a = a[0]
        print(a,type(a))
        if a == 1.0:
            print("Yes")
        elif a== 0:
            print("No")
        self.result.append(a)
    
        l.append(a)    
        self.updateDataset(l)
        
    def updateDataset(self,li):
        self.dataset = self.dataset.append(pd.DataFrame([li],columns = self.col))
        print(self.dataset.tail())
        self.dataset_update()
        
        
ms = ModelSelection()
ms.SVM()
ms.RandomForest()
```
